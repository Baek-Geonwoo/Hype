```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./data/result.csv')
data
```

```python
raw_data = pd.read_csv('./data/raw_data.csv')
raw_data
```

```python
add_lst = ['행정동명', '행정동코드', '상호명', '경도', '위도']
data.loc[:, add_lst] = raw_data[add_lst]
data
```

```python
data = data[['행정동명', '행정동코드', '업종', '상호명', '경도', '위도', '상권']]
data
```

```python
area = [0]*len(data['행정동코드'].unique())

for i, a in enumerate(sorted(data['행정동코드'].unique())):
    area[i] = data.loc[data['행정동코드'] == a, :]
area[0]
```

```python
cnt = 0
typo_lst = []
for a in range(len(area)):
    if len(area[a]['행정동명'].unique()) != 1:
        typo_lst.append(Counter(area[a]['행정동명']).keys())
typo_lst
```

```python
for name, typo in typo_lst:
    data.loc[data['행정동명'] == typo, '행정동명'] = name

area = [0]*len(data['행정동코드'].unique())

for i, a in enumerate(sorted(data['행정동코드'].unique())):
    area[i] = data.loc[data['행정동코드'] == a, :]
    
for a in range(len(area)):
    if len(area[a]['행정동명'].unique()) != 1:
        print('error!')
```

```python
area[0]
```

```python
area_counter = [
    list(area[a].iloc[0,:2])
    + np.array(Counter(area[a]['업종']).most_common(3)).flatten().tolist()
    for a in range(len(area))
]
area_counter[0]
```

```python
sample = pd.DataFrame(
    columns=[
        '행정동명', '행정동코드',
        'Top1', 'number of Top1',
        'Top2', 'number of Top2',
        'Top3', 'number of Top3'
    ]
)
sample
```

```python
for i in range(len(area_counter)):
    sample = sample.append(
        {k:v for k,v in zip(sample.columns, area_counter[i])},
        ignore_index=True
    )
sample
```

```python
sample.to_csv('./data/sample.csv', index=False)
```
