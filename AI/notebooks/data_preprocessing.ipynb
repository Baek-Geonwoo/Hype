```python
meaningful_cols = [
    '지점명', '상권업종대분류코드', '상권업종중분류코드', '상권업종소분류코드', '표준산업분류코드',
    '행정동코드', '대지구분코드', '건물관리번호', '신우편번호', '경도', '위도'
] # 대지구분코드 건물관리번호, 신우편번호는 애매함
```

```python
clear_data = data[meaningful_cols]
```

```python
clear_data.dtypes
```

```python
clear_data.isna().sum()
```

# 결측치 제거
* 지점명의 경우 지점유무 컬럼으로 변경하여 NULL인 경우 지점이 없는 것으로 One-Hot-Encoding한다.
* 표준산업분류코드, 신우편번호 : 각각 상권업종분류코드, 건물관리번호로 대체가 가능하므로 삭제한다.

```python
clear_data
```

```python
clear_data['지점유무'] = np.where(clear_data['지점명'].isna(), 0, 1)
clear_data.drop('지점명', axis=1, inplace=True)
```

```python
clear_data
```

# 필요없는 컬럼 제거
표준산업분류코드, 지번코드, 신우편번호 : 건물관리번호, 경도, 위도와 의미가 중복되므로 삭제
상권업종대분류코드, 상권업종중분류코드 : 상권업종소분류코드에 포함되고, 상권업종소분류코드가 부분집합이므로 삭제

```python
clear_data.drop(['상권업종대분류코드', '상권업종중분류코드', '표준산업분류코드', '신우편번호'], axis=1, inplace = True)
```

# 컬럼 가공
상권업종소분류코드 : 상권업종소분류코드의 앞 3자리는 상권업종대분류코드와 상권업종중분류코드이므로 제거

```python
clear_data['상권업종소분류코드'] = clear_data['상권업종소분류코드'].str[3:]
```

# 분류형 데이터 수치형으로 변환
* 상권업종소분류코드 : 상권업종소분류코드의 앞자리는 알파벳 A, B이므로 16진수로 간주하고 10진수로 변환하여 label encoding하여 수치형으로 변환한다.
* 건물관리번호 : 건물관리번호 컬럼은 object형이나 수치형데이터이므로 형변환한다.

```python
clear_data['상권업종소분류코드'].str[0].unique()
```

```python
clear_data['상권업종소분류코드'] = clear_data['상권업종소분류코드'].apply(int, base=16)
```

```python
clear_data['건물관리번호'] = clear_data['건물관리번호'].astype(float)
```

# 데이터 저장
* 전처리가 끝난 데이터를 디스크에 저장한다.

```python
clear_data.to_csv('./data/train_data.csv', index=False)
```
